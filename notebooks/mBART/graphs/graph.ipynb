```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Setting the font to Times New Roman
rcParams['font.family'] = 'sans-serif'
```

```python
df1 = pd.read_csv('./after/train_loss.csv')
df2 = pd.read_csv('./after/validation_loss.csv')
```

```python
# df2
```

```python
# df1
```

```python
plt.figure(figsize=(6, 4))

# Plotting df1 with markers 'o'
plt.plot(df1['Step'], df1['loss'], linestyle='-.', color='k', label='Train Loss')

# Plotting df2 with markers 's' (square markers)
# plt.plot(df2['Step'], df2['loss'], linestyle='--', color='b', label='df2 Loss')

plt.title('Train Loss vs Steps')
plt.xlabel('Steps')
plt.ylabel('Train Loss')

# Adding a legend to differentiate the plots
plt.legend()

# Removing the grid as per the instruction
plt.grid(False)

# Show the plot
plt.savefig('train_loss.png')
plt.show()
```

```python
plt.figure(figsize=(6, 4))

# Plotting df1 with markers 'o'
# plt.plot(df1.index, df1['train_loss'], linestyle='-.', color='k', label='df1 Train Loss')

# Plotting df2 with markers 's' (square markers)
plt.plot(df2['Step'], df2['loss'], linestyle='-', color='k', label='Validation Loss')

plt.title('Validation Loss vs Steps')
plt.xlabel('Steps')
plt.ylabel('Validation Loss')

# Adding a legend to differentiate the plots
plt.legend()

# Removing the grid as per the instruction
plt.grid(False)

# Show the plot
plt.savefig('validation_loss.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creating the figure
plt.figure(figsize=(8, 6))

# Plotting the train loss with distinct markers and line style
plt.plot(df1['Step'], df1['loss'], linestyle='-', color='k', label='Train Loss')

# Plotting the validation loss with distinct markers and line style
plt.plot(df2['Step'], df2['loss'], linestyle='-.', color='k', label='Validation Loss')

# Adding title and labels
plt.title('Train and Validation Loss vs Steps')
plt.xlabel('Steps')
plt.ylabel('Loss')

# Adding a legend to differentiate the curves
plt.legend()

# Removing the grid for cleaner visualization
plt.grid(False)

# Saving the figure to a file
plt.savefig('train_validation_loss.png')

# Displaying the plot
plt.show()
```
