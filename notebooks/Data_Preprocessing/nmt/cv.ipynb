```python
import os
import pandas as pd
from pydub import AudioSegment
from pydub.utils import which

# Ensure pydub can find ffmpeg and ffprobe
AudioSegment.converter = which("ffmpeg")
AudioSegment.ffprobe = which("ffprobe")

def convert_mp3_to_wav(df, output_dir='./wav_files/'):
    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    for index, mp3_path in enumerate(df):
        print(mp3_path)
        try:
            # Load the .mp3 file
            audio = AudioSegment.from_mp3(mp3_path)
            
            # Construct the output file path
            wav_path = os.path.join(output_dir, os.path.basename(mp3_path).replace('.mp3', '.wav'))
            
            # Export the audio file as .wav
            audio.export(wav_path, format="wav")
            print(f"Converted {mp3_path} to {wav_path}")
        except Exception as e:
            print(f"Failed to convert {mp3_path}: {e}")

# Example usage
convert_mp3_to_wav(df['Audio'].iloc[2739:], output_dir='./wav_files/')
```
