```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Setting the font to Times New Roman
rcParams['font.family'] = 'sans-serif'
```

```python
df1 = pd.read_csv('loss.csv')
```

```python
import matplotlib.pyplot as plt

# Updated figure size
plt.figure(figsize=(8, 5))

# Corrected plotting for df1 with markers 'o'
plt.plot(df1[df1['Step'] % 50 == 0]['Step'], df1[df1['Step'] % 50 == 0]['loss'], linestyle='-.', color='k', marker='o', label='Train Loss')

plt.title('Train Loss vs Steps')
plt.xlabel('Steps')
plt.ylabel('Loss')

# Adding a legend
plt.legend()

# Grid is explicitly removed
plt.grid(False)

# Save and show the plot
plt.savefig('train_loss.png')
plt.show()
```
