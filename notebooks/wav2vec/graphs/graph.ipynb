```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Setting the font to Times New Roman
rcParams['font.family'] = 'serif'
```

```python
df1 = pd.read_csv('train.csv')
df2 = pd.read_csv('validation.csv')
```

```python
import matplotlib.pyplot as plt

# Creating the figure
plt.figure(figsize=(8, 6))

# Plotting the train loss with distinct markers and line style
plt.plot(df1['step'], df1['loss'], linestyle='-', color='k', label='Train Loss')

# Plotting the validation loss with distinct markers and line style
plt.plot(df2['step'], df2['loss'], linestyle='-.', color='k', label='Validation Loss')

# Adding title and labels
plt.title('Train and Validation Loss vs Steps')
plt.xlabel('Steps')
plt.ylabel('Loss')

# Adding a legend to differentiate the curves
plt.legend()

# Removing the grid for cleaner visualization
plt.grid(False)

# Saving the figure to a file
plt.savefig('train_validation_loss.png')

# Displaying the plot
plt.show()
```

```python
df3 = pd.read_csv('wer.csv')
```

```python
import matplotlib.pyplot as plt


# Creating the figure
plt.figure(figsize=(8, 6))

# Plotting the BLEU score
plt.plot(df3['Step'], df3['wer'], linestyle='-', color='k', label='WER Score')

# Adding title and labels
plt.title('WER Score vs Steps', fontsize=14)
plt.xlabel('Steps', fontsize=12)
plt.ylabel('WER Score', fontsize=12)

# Adding a legend to differentiate the curves
plt.legend(fontsize=10)

# Removing the grid for cleaner visualization
plt.grid(False)

# Saving the figure to a file
plt.savefig('wer_times_new_roman.png')

# Displaying the plot
plt.show()
```
